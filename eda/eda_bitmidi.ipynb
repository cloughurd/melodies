```python
import os
import mido
import pretty_midi
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
import mido
import pretty_midi
import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt

# Directory with MIDI files
midi_folder = "../data/"
midi_files = [f for f in os.listdir(midi_folder) if f.endswith(".mid")]

# Initialize stats
tempo_list = []
note_counts = []
track_counts = []
pitch_ranges = []
msg_type_counts = Counter()
common_notes = Counter()

# Analyze a sample of MIDI files
for filename in midi_files[:50]:  # Limiting to 50 for speed
    try:
        path = os.path.join(midi_folder, filename)
        mid = mido.MidiFile(path)
        pm = pretty_midi.PrettyMIDI(path)

        # Tempo and notes
        try:
            tempo = pm.estimate_tempo()
            tempo_list.append(tempo)
        except:
            pass
        
        # Pitch range
        pitches = [note.pitch for inst in pm.instruments for note in inst.notes]
        if pitches:
            pitch_ranges.append(max(pitches) - min(pitches))
            common_notes.update(pitches)
            note_counts.append(len(pitches))
        
        # Track count
        track_counts.append(len(mid.tracks))

        # Message types
        for track in mid.tracks:
            for msg in track:
                msg_type_counts[msg.type] += 1

    except Exception as e:
        print(f"Error processing {filename}: {e}")

# Convert stats to summaries
avg_tempo = np.mean(tempo_list)
avg_note_count = np.mean(note_counts)
avg_tracks = np.mean(track_counts)
avg_pitch_range = np.mean(pitch_ranges)

# Most common notes (pitch numbers)
most_common_notes = common_notes.most_common(10)
msg_type_df = pd.DataFrame.from_dict(msg_type_counts, orient='index', columns=['count']).sort_values(by='count', ascending=False)

print("Message Type Counts:\n", msg_type_df)

# Generate plots
plt.figure(figsize=(10, 5))
plt.bar(*zip(*most_common_notes))
plt.title("Top 10 Most Common Notes")
plt.xlabel("MIDI Pitch")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(10, 5))
plt.hist(tempo_list, bins=20, color='skyblue', edgecolor='black')
plt.title("Tempo Distribution")
plt.xlabel("Tempo (BPM)")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(10, 5))
plt.hist(note_counts, bins=20, color='lightgreen', edgecolor='black')
plt.title("Note Count per File")
plt.xlabel("Number of Notes")
plt.ylabel("Count")
plt.show()
```
